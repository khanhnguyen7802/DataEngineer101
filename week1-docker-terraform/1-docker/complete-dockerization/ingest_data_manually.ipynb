# This file is to ingest the data (.csv file) to Postgres

## Download dataset

```python
import subprocess
subprocess.run(["wget", "https://github.com/DataTalksClub/nyc-tlc-data/releases/download/green/green_tripdata_2019-10.csv.gz"])
subprocess.run(["wget", "https://github.com/DataTalksClub/nyc-tlc-data/releases/download/misc/taxi_zone_lookup.csv"])
```

```python
import gzip
import shutil

with gzip.open('./green_tripdata_2019-10.csv.gz', 'rb') as f_in:
    with open('./green_tripdata_2019-10.csv', 'wb') as f_out:
        shutil.copyfileobj(f_in, f_out)
```

```python
subprocess.run(["rm", "green_tripdata_2019-10.csv.gz"])
```

## Ingest `green_tripdata_2019-10.csv` file 

```python
import pandas as pd
```

```python
trips = pd.read_csv("./green_tripdata_2019-10.csv")
```

```python
# trips
```

```python
print(pd.io.sql.get_schema(trips, name="green_trip_data"))
```

```python
trips.lpep_pickup_datetime = pd.to_datetime(trips.lpep_pickup_datetime) 
trips.lpep_dropoff_datetime = pd.to_datetime(trips.lpep_dropoff_datetime)
```

```python
print(pd.io.sql.get_schema(trips, name="green_trip_data"))
```

```python
from sqlalchemy import create_engine
```

```python
# create_engine('postgresql+psycopg2://user:password@hostname/database_name')
engine = create_engine('postgresql://root:root@pgdatabase:5432/green_trip')
```

```python
engine.connect()
```

```python
print(pd.io.sql.get_schema(trips, name="green_trip_data", con=engine))
```

```python
from time import time 
```

```python
df_iter = pd.read_csv("./green_tripdata_2019-10.csv", iterator=True, chunksize=50000)
i = 0

for chunk in df_iter:
  t_start = time()
  
  chunk.lpep_pickup_datetime = pd.to_datetime(chunk.lpep_pickup_datetime) 
  chunk.lpep_dropoff_datetime = pd.to_datetime(chunk.lpep_dropoff_datetime)

  chunk.to_sql(name='green_trip_data', con=engine, if_exists='append')

  t_end = time()


  print(f"inserted chunk {i}: took {t_end - t_start:.2f} seconds")
  i += 1
```
