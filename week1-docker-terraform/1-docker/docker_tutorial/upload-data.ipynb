```python
import pyarrow.parquet as pq
import pandas as pd
```

```python
trips = pq.read_table('yellow_tripdata_2024-01.parquet')
```

```python
trips = trips.to_pandas()
```

```python
trips
```

```python
print(pd.io.sql.get_schema(trips, name="yellow_taxi_data"))
```

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://root:root@localhost:5434/ny_taxi')
```

```python
engine.connect()
```

```python
print(pd.io.sql.get_schema(trips, name="yellow_taxi_data", con=engine))
```

```python
from time import time 
```

```python
parquet_file = pq.ParquetFile('yellow_tripdata_2024-01.parquet')

for batch in parquet_file.iter_batches():
    t_start = time()
    batch_df = batch.to_pandas()
    batch_df.to_sql(name='yellow_taxi_data', 
                    con=engine, if_exists='append')
    t_end=time()
    print(f"inserted another batch..., took {t_end - t_start:.2f} seconds")

    break # just take a small chunk of the database due to storage size 
```
