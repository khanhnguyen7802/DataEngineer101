### Ingest `green_tripdata_2025-11.parquet` file 

```python
import pandas as pd
import os 
```

```python
# create an empty folder to contain the data
os.makedirs(f"{os.getcwd()}/green_trip_data", exist_ok=True)
```

```python
trips = pd.read_parquet("green_tripdata_2025-11.parquet")
```

```python
trips
```

```python
print(pd.io.sql.get_schema(trips, name="green_trip_2025-11", con=None))
```

```python
from sqlalchemy import create_engine
```

```python
# create database engine, saved for later connection 

# create_engine('postgresql+psycopg2://user:password@hostname/database_name')
engine = create_engine('postgresql://root:root@localhost:5434/green_trip')
```

```python
engine.connect()
```

```python
# check again the schema, now with the engine
print(pd.io.sql.get_schema(trips, name="green_trip_2025-11", con=engine))
```

```python
df = pd.read_parquet("green_tripdata_2025-11.parquet")
df.to_sql(
    name="green_trip_2025-11",
    con=engine,
    if_exists="append",
    index=False
)
```

### Ingest `taxi_zone_lookup.csv` file

```python
zones = pd.read_csv("taxi_zone_lookup.csv")
```

```python
zones
```

```python
print(pd.io.sql.get_schema(zones, name="zone_data"))
```

```python
from sqlalchemy import create_engine
```

```python
# create_engine('postgresql+psycopg2://user:password@hostname/database_name')
engine = create_engine('postgresql://root:root@localhost:5434/green_trip')
```

```python
engine.connect()
```

```python
print(pd.io.sql.get_schema(zones, name="zone_data", con=engine))
```

```python
from time import time 
```

```python
df_iter = pd.read_csv("taxi_zone_lookup.csv", iterator=True)
i = 0

for chunk in df_iter:
  t_start = time()
  
  chunk.to_sql(name='zone_data', con=engine, if_exists='append')

  t_end = time()


  print(f"inserted chunk {i}: took {t_end - t_start:.2f} seconds")
  i += 1
```
